# Spherical Harmonics
- $Y_{00} \propto 1$, $Y_{20} \propto \cos^2\theta$


- What do these look like?

```python
xs = np.linspace(0, np.pi, 100)
y00 = np.ones_like(xs) + .1
y20 = np.cos(xs)**2
```

```python
plt.plot(xs, y00, label=r'$Y_{00}$')
plt.plot(xs, y20, label=r'$Y_{20}$')
plt.xlabel(r'$\theta$')
plt.ylabel('y')
plt.legend();
```

- What happens if we limit $\theta$ to $\pm 10$ deg?

```python
m1 = xs <= np.radians(10)
m2 = xs >= np.radians(170)
```

```python
plt.plot(xs[m1], y00[m1], c='blue', label=r'$Y_{00}$')
plt.plot(xs[m2], y00[m2], c='blue',)
plt.plot(xs[(~m1) & (~m2)], y00[(~m1) & (~m2)], c='gray', linestyle='--')
plt.plot(xs[m1], y20[m1], c='red', label=r'$Y_{20}$')
plt.plot(xs[m2], y20[m2], c='red',)
plt.plot(xs[(~m1) & (~m2)], y20[(~m1) & (~m2)], c='gray', linestyle='--')
plt.xlabel(r'$\theta$')
plt.ylabel('y')

plt.legend();
```

- These two terms very hard to distinguish, so linear parameters are highly correlated

$$ V = c_{20} Y_{20} + c_{00} Y_{00}$$
